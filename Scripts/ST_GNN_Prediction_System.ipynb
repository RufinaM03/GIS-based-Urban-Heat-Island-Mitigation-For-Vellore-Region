```python
# Import libraries

import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv, GraphConv, global_mean_pool

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Check device (CPU in your case)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# -------------------- Imports --------------------
import geopandas as gpd
import pandas as pd
import torch
from pathlib import Path

# -------------------- File Paths --------------------
WARD_SHP = r"Vellore Shapefile/Vellore_boundary.shp"                    # ward polygon shapefile
TIMESERIES_CSV = "Preprocessed_data/vellore_polygons_climate_2019_2024.xlsx"  # climate time series
STATIC_FEATURES_CSV = "Preprocessed_data/Spatial_Features_Extracted.csv"      # static features

# -------------------- Columns --------------------
WARD_COL = "NAME_3"       # ward names in shapefile
WARD_ID_COL = "area_name" # ward id in time series and static data
DATE_COL = "date"         # actual column in time series
TARGET_COL = "T2M"        # temperature as proxy for UHI
FEATURE_COLS = ["RH2M"]   # additional features (humidity)

# -------------------- Training hyperparams --------------------
HIDDEN_CHANNELS = 64
LEARNING_RATE = 0.005
EPOCHS = 2000
TRAIN_RATIO = 0.8   # fraction of time steps for training
PATIENCE = 50   # stop if no improvement for 50 epochs
BEST_VAL_LOSS = float("inf")
patience_counter = 0
best_model_wts = None

# -------------------- Device --------------------
DEVICE = torch.device("cpu")
print("Running only on CPU")

# -------------------- Read Shapefile --------------------
wards = gpd.read_file(WARD_SHP)
if wards.crs is None:
    wards = wards.set_crs(epsg=4326)

# Check column
if WARD_COL not in wards.columns:
    raise ValueError(f"Shapefile must contain '{WARD_COL}' column")

# Rename ward column to match time series
wards = wards.rename(columns={WARD_COL: WARD_ID_COL})

# -------------------- Read Time Series --------------------
if TIMESERIES_CSV.endswith(".xlsx"):
    ts = pd.read_excel(TIMESERIES_CSV)
else:
    ts = pd.read_csv(TIMESERIES_CSV)

# Check required columns
required_ts_cols = [WARD_ID_COL, DATE_COL, TARGET_COL] + FEATURE_COLS
for col in required_ts_cols:
    if col not in ts.columns:
        raise ValueError(f"Time series file must contain '{col}' column")

# Convert date column to datetime, coerce errors
ts[DATE_COL] = pd.to_datetime(ts[DATE_COL], errors='coerce')

# Drop rows where date could not be parsed
num_invalid_dates = ts[DATE_COL].isnull().sum()
if num_invalid_dates > 0:
    print(f"Dropping {num_invalid_dates} rows with invalid dates")
    ts = ts.dropna(subset=[DATE_COL])

# -------------------- Read Static Features --------------------
if Path(STATIC_FEATURES_CSV).exists():
    if STATIC_FEATURES_CSV.endswith(".xlsx"):
        static_df = pd.read_excel(STATIC_FEATURES_CSV)
    else:
        static_df = pd.read_csv(STATIC_FEATURES_CSV)

    if WARD_ID_COL not in static_df.columns:
        raise ValueError(f"Static features must contain '{WARD_ID_COL}' column")
else:
    static_df = None

# -------------------- Merge Data --------------------
# Merge static features into wards
if static_df is not None:
    wards_static = wards.merge(static_df, on=WARD_ID_COL, how="left")
else:
    wards_static = wards.copy()

# Merge time series with wards + static features
full_df = ts.merge(wards_static, on=WARD_ID_COL, how="left")

# -------------------- Engineer UHI Proxy --------------------
# Simple UHI proxy: deviation from daily mean temperature
full_df["UHI_proxy"] = full_df[TARGET_COL] - full_df.groupby(DATE_COL)[TARGET_COL].transform("mean")

# -------------------- Convert to GeoDataFrame --------------------
full_gdf = gpd.GeoDataFrame(full_df, geometry="geometry", crs=wards.crs)

# -------------------- Debug Output --------------------
print("\nWards Columns:", wards.columns.tolist())
print("Time Series Columns:", ts.columns.tolist())
if static_df is not None:
    print("Static Features Columns:", static_df.columns.tolist())
print("\nSample Data with UHI Proxy:")
print(full_gdf[[WARD_ID_COL, DATE_COL, TARGET_COL, "UHI_proxy"]].head())
```

```python
# -------------------- Build Adjacency Graph (CPU only) --------------------
# Use wards from the merged full_gdf (geometry + ward_id)
wards_gdf = full_gdf.drop_duplicates(subset=[WARD_ID_COL]).set_index(WARD_ID_COL)
ward_ids = list(wards_gdf.index)
n_nodes = len(ward_ids)
print("\nNumber of wards (nodes):", n_nodes)

# Build spatial index for faster neighbor search
sindex = wards_gdf.sindex

# Build adjacency list
edges = []
for i, id_i in enumerate(ward_ids):
    geom_i = wards_gdf.loc[id_i].geometry

    # Candidate neighbors using bounding boxes
    possible_idxs = list(sindex.intersection(geom_i.bounds))
    possibles = wards_gdf.iloc[possible_idxs]

    for j_id, row in possibles.iterrows():
        if j_id == id_i:
            continue
        # adjacency: touches or intersects
        if geom_i.touches(row.geometry) or geom_i.intersects(row.geometry):
            u = i
            v = ward_ids.index(j_id)
            edges.append((u, v))

# Make unique undirected edges
edges_unique = set()
for u, v in edges:
    if u == v:
        continue
    a, b = min(u, v), max(u, v)
    edges_unique.add((a, b))

edges_list = list(edges_unique)
print("Number of adjacency edges:", len(edges_list))
if len(edges_list) == 0:
    raise ValueError("No adjacency edges found â€” check your shapefile geometry/CRS")

# -------------------- Example: adjacency dictionary --------------------
adj_dict = {i: [] for i in range(n_nodes)}
for u, v in edges_list:
    adj_dict[u].append(v)
    adj_dict[v].append(u)

print("\nSample adjacency (first 5 wards):")
for i in range(min(5, n_nodes)):
    print(f"Ward {ward_ids[i]} neighbors -> {[ward_ids[j] for j in adj_dict[i]]}")

# Explicitly confirm CPU usage
print("\nAdjacency graph built.")
```

```python
import numpy as np
import torch

# -------------------- Build edge_index and edge_weight --------------------
# Create edge_index (2 x E) for undirected graph
edge_index = torch.tensor(
    [[e[0] for e in edges_list] + [e[1] for e in edges_list],
     [e[1] for e in edges_list] + [e[0] for e in edges_list]],
    dtype=torch.long
).to("cpu")   # force CPU
print("Edges count (undirected):", len(edges_list))

# Edge weights (all ones for now)
edge_weight = torch.ones(edge_index.shape[1], dtype=torch.float).to("cpu")

# -------------------- Prepare time-series node features and targets --------------------
# Ensure every ward has entries for each date; create full date index
all_dates = sorted(ts[DATE_COL].unique())
T = len(all_dates)
print("Time steps:", T)

# Mapping from ward_id to index
ward_to_idx = {wid: i for i, wid in enumerate(ward_ids)}

# Prepare static features per node
if static_df is not None:
    static_df = static_df.set_index(WARD_ID_COL).reindex(ward_ids).fillna(0)
    static_features = static_df.values.astype(float)  # shape (n_nodes, n_static_feats)
    static_feat_names = list(static_df.columns)
else:
    # fallback: extract numeric columns from wards shapefile if available
    possible_cols = [c for c in wards.columns if c.lower() in
                     ("ndvi_mean","ndwi_mean","ndbi_mean","sand_pct","oc","population")]
    if len(possible_cols) > 0:
        static_features = wards[possible_cols].fillna(0).values.astype(float)
        static_feat_names = possible_cols
    else:
        static_features = np.zeros((n_nodes, 1))
        static_feat_names = ["bias"]

print("Static features shape:", static_features.shape, "columns:", static_feat_names)

# -------------------- Build time-series feature matrix X_t and target y_t --------------------
# Features: static + time-varying (e.g., RH2M, temp/UHI)
n_features = static_features.shape[1] + len(FEATURE_COLS)
X_t = np.zeros((T, n_nodes, n_features), dtype=float)
y_t = np.zeros((T, n_nodes), dtype=float)

# Fill in features and target
for t_idx, date in enumerate(all_dates):
    date_df = full_gdf[full_gdf[DATE_COL] == date]
    for _, row in date_df.iterrows():
        ward_idx = ward_to_idx[row[WARD_ID_COL]]
        # Static features
        X_t[t_idx, ward_idx, :static_features.shape[1]] = static_features[ward_idx]
        # Time-varying features
        for f_idx, f_col in enumerate(FEATURE_COLS):
            X_t[t_idx, ward_idx, static_features.shape[1] + f_idx] = row[f_col]
        # Target: UHI proxy
        y_t[t_idx, ward_idx] = row["UHI_proxy"]

# Convert to torch tensors (CPU only)
X_t = torch.tensor(X_t, dtype=torch.float, device="cpu")  # shape: (T, n_nodes, n_features)
y_t = torch.tensor(y_t, dtype=torch.float, device="cpu")  # shape: (T, n_nodes)

print("Time-series features X_t shape:", X_t.shape)
print("Time-series targets y_t shape:", y_t.shape)
print("Data prepared")
```

```python
# -------------------- Build time-series arrays (CPU only) --------------------
# Combine static + temporal features (T2M/UHI proxy already in y_t if needed)
temporal_feat_cols = [TARGET_COL] + FEATURE_COLS  # e.g., ["T2M", "RH2M"]
if len(temporal_feat_cols) == 0:
    raise ValueError("Time series must contain at least one temporal column like T2M or RH2M")

X_t_list = []  # list of numpy arrays (n_nodes x num_features)
y_t_list = []  # list of numpy arrays (n_nodes x 1)

# Preindex ts by (ward_id, date) for fast lookup
ts_map = ts.set_index([WARD_ID_COL, DATE_COL])

for date in all_dates:
    node_feats = []
    node_targets = []
    for wid in ward_ids:
        # temporal features
        try:
            row = ts_map.loc[(wid, date)]
            if isinstance(row, pd.DataFrame):
                row = row.mean()  # if multiple rows, take mean
            temporal_vals = row[temporal_feat_cols].values.astype(float)
        except KeyError:
            temporal_vals = np.array([np.nan] * len(temporal_feat_cols), dtype=float)

        # target: UHI proxy
        try:
            tgt = float(
                full_gdf[
                    (full_gdf[WARD_ID_COL] == wid) & (full_gdf[DATE_COL] == date)
                ]["UHI_proxy"].values[0]
            )
        except Exception:
            tgt = np.nan

        # combine static + temporal
        combined = np.concatenate([static_features[ward_to_idx[wid]], temporal_vals])
        node_feats.append(combined)
        node_targets.append([tgt])

    X_t_list.append(np.asarray(node_feats))     # shape (n_nodes, n_features)
    y_t_list.append(np.asarray(node_targets))   # shape (n_nodes, 1)

# Convert to numpy arrays
X_t = np.asarray(X_t_list)  # shape (T, n_nodes, n_features)
y_t = np.asarray(y_t_list)  # shape (T, n_nodes, 1)

# Convert to torch tensors on CPU
X_t = torch.tensor(X_t, dtype=torch.float, device="cpu")
y_t = torch.tensor(y_t, dtype=torch.float, device="cpu")

print("\nRaw time-series shapes: X_t", X_t.shape, "y_t", y_t.shape)
print("Time-series arrays prepared on CPU")
```

```python
from sklearn.preprocessing import StandardScaler
import torch
import numpy as np

# -------------------- Handle missing values in features --------------------
for node in range(X_t.shape[1]): # Iterate through nodes
    for feat in range(X_t.shape[2]): # Iterate through features
        series = X_t[:, node, feat] # Get the time series for this node and feature

        if torch.all(torch.isnan(series)): # Use torch.all and torch.isnan
            series[:] = 0.0
        else:
            mask = torch.isnan(series) # Use torch.isnan
            if mask.any():
                # Forward fill
                for i in range(1, len(series)):
                    if torch.isnan(series[i]) and not torch.isnan(series[i-1]):
                        series[i] = series[i-1]
                # Backward fill
                for i in range(len(series)-2, -1, -1):
                    if torch.isnan(series[i]) and not torch.isnan(series[i+1]):
                        series[i] = series[i+1]
                # Remaining nans -> 0
                series[torch.isnan(series)] = 0.0 # Use torch.isnan

        X_t[:, node, feat] = series

# -------------------- Handle missing values in target --------------------
for node in range(y_t.shape[1]): # Iterate through nodes
    series = y_t[:, node, 0] # Get the time series for this node

    if torch.all(torch.isnan(series)): # Use torch.all and torch.isnan
        series[:] = 0.0
    else:
        nans = torch.isnan(series) # Use torch.isnan
        if nans.any():
            # Linear interpolation using numpy for simplicity after converting to numpy
            series_np = series.numpy()
            not_nans_np = ~np.isnan(series_np)
            xs_np = np.arange(len(series_np))
            series_np[np.isnan(series_np)] = np.interp(xs_np[np.isnan(series_np)], xs_np[not_nans_np], series_np[not_nans_np])
            series = torch.tensor(series_np, dtype=torch.float, device="cpu") # Convert back to torch tensor

    y_t[:, node, 0] = series

# -------------------- Standardize features --------------------
num_features = X_t.shape[2]
scaler = StandardScaler()

# Reshape, scale, and reshape back
# Convert to numpy for StandardScaler
X_t_numpy = X_t.reshape(-1, num_features).numpy()
X_t_flat_scaled = scaler.fit_transform(X_t_numpy)
X_t = torch.tensor(X_t_flat_scaled.reshape(X_t.shape), dtype=torch.float, device="cpu")


# -------------------- Print result --------------------
print("\nPrepared time-series data:")
print("X_t shape:", X_t.shape)
print("y_t shape:", y_t.shape)
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch_geometric_temporal.signal import StaticGraphTemporalSignal
from torch_geometric_temporal.nn.recurrent import TGCN
from torch.optim.lr_scheduler import ReduceLROnPlateau # Import ReduceLROnPlateau
import copy # Import copy for saving best model weights

# -------------------- Force CPU --------------------
DEVICE = torch.device("cpu")
print("Using device:", DEVICE)

# -------------------- Dataset preparation --------------------

# Build edge_index (CPU)
edge_index = torch.tensor(
    [[e[0] for e in edges_list] + [e[1] for e in edges_list],
     [e[1] for e in edges_list] + [e[0] for e in edges_list]],
    dtype=torch.long,
    device="cpu"
)
print("Edges count (undirected):", len(edges_list))

# Edge weights (CPU)
edge_weight = torch.ones(edge_index.shape[1], dtype=torch.float, device="cpu")

# Convert features/targets to numpy float32
X_t_np = X_t.cpu().numpy().astype(np.float32)   # shape (T, N, F)
y_t_np = y_t.cpu().numpy().astype(np.float32).reshape(y_t.shape[0], y_t.shape[1])  # (T, N)

# Build StaticGraphTemporalSignal with numpy arrays
dataset = StaticGraphTemporalSignal(
    edge_index=edge_index.numpy(),
    edge_weight=edge_weight.numpy(),
    features=[X_t_np[t] for t in range(X_t_np.shape[0])],
    targets=[y_t_np[t] for t in range(y_t_np.shape[0])]
)

# Convert iterator to list for indexing and len()
dataset = list(dataset)

train_steps = int(len(dataset) * TRAIN_RATIO)
test_steps = len(dataset) - train_steps
train_dataset = list(dataset)[:train_steps]
test_dataset = list(dataset)[train_steps:]

print("Train steps:", train_steps, "Test steps:", test_steps)

# -------------------- Model --------------------
class TGCNNet(nn.Module):
    def __init__(self, node_features, hidden_channels):
        super(TGCNNet, self).__init__()
        self.tgcn = TGCN(node_features, hidden_channels)
        self.linear = nn.Linear(hidden_channels, 1)

    def forward(self, x, edge_index, edge_weight):
        h = self.tgcn(x, edge_index, edge_weight)
        out = self.linear(h)
        return out.squeeze(-1)

model = TGCNNet(node_features=X_t_np.shape[2], hidden_channels=HIDDEN_CHANNELS).to("cpu")
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=20) # Remove verbose=True
criterion = nn.MSELoss()

# -------------------- Training loop --------------------
print("\nStart training...")

BEST_VAL_LOSS = float("inf") # Initialize BEST_VAL_LOSS
patience_counter = 0 # Initialize patience_counter
best_model_wts = None # Initialize best_model_wts
PATIENCE = 50   # stop if no improvement for 50 epochs


for epoch in range(1, EPOCHS+1):
    model.train()
    epoch_losses = []

    for snapshot in train_dataset:
        x = snapshot.x.to("cpu")
        y = snapshot.y.to("cpu")
        edge_idx = snapshot.edge_index.to("cpu")
        edge_w = snapshot.edge_weight

        if edge_w is None:  # fallback: all ones
          edge_w = torch.ones(edge_idx.shape[1], dtype=torch.float, device="cpu")

        optimizer.zero_grad()
        out = model(x, edge_idx, edge_w)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        epoch_losses.append(loss.item())

    val_loss = np.mean(epoch_losses)

    # --- Scheduler step ---
    scheduler.step(val_loss)

    if epoch % 10 == 0 or epoch == 1:
        print(f"Epoch {epoch:03d}, Train Loss: {np.mean(epoch_losses):.4f}")

        # --- Early stopping check ---
    if val_loss < BEST_VAL_LOSS:
        BEST_VAL_LOSS = val_loss
        patience_counter = 0
        best_model_wts = copy.deepcopy(model.state_dict())
    else:
        patience_counter += 1

    if patience_counter >= PATIENCE:
        print(f"\nEarly stopping at epoch {epoch}")
        break

# Load best weights
if best_model_wts:
    model.load_state_dict(best_model_wts)
print("\nTraining completed.")

# -------------------- Evaluation --------------------
model.eval()
test_losses = []
with torch.no_grad():
    for snapshot in test_dataset:
        x = snapshot.x.to("cpu")
        y = snapshot.y.to("cpu")
        edge_idx = snapshot.edge_index.to("cpu")
        edge_w = snapshot.edge_weight

        out = model(x, edge_idx, edge_w)
        loss = criterion(out, y)
        test_losses.append(loss.item())

print(f"Test MSE: {np.mean(test_losses):.4f}")
```
